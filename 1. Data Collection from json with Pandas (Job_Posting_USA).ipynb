```python
# 1) import pandas and the dataset extracted from the link below

import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

df = pd.read_json(url)
```

```python
df.info()
df
```

```python
# 2) replace ',' with ';'

df = df.replace(',',';', regex = True)
```

```python
# 3) number of jobs for the Python technology

df[['Role']][df['Key Skills'].str.contains('Python')].count()
```

```python
# 4) number of jobs in the locations of USA

df.Location.value_counts()
```

```python
# 5) list of all job locations where Python is used

locations_python = df[['Location','Role','Key Skills']][df['Key Skills'].str.contains('Python')].Location.unique()
locations_python
```

```python
# 6) import the DataFrame to csv file

df.to_csv('Job_Posting_USA.csv', index = False)
```

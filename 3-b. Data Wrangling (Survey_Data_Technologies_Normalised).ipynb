```python
# 1) import pandas

import pandas as pd

# import df from csv
df = pd.read_csv('survey_data_technologies_not_cleaned.csv')
```

```python
# 5) remove empty coloumns

df = df.iloc[:,:10]
df
```

```python
# 6) remove empty rows (only nan) and reset index

df = df.dropna(how = 'all').reset_index(drop = True)
df
```

```python
# 7) convert the data type of the column 'Respondent'

df.Respondent = df.Respondent.astype(int)
df.info()
df.sample()
```

```python
# 8) find duplicates

df[df.duplicated()].shape[0]
```

```python
# 9) find count and % of null in dataset

if df.isna().sum().sum():
    for column in df.columns:
        if df[column].isna().sum() > 0:
            count_null = df[column].isna().sum()
            percent_null = round(count_null/df.shape[0]*100,2)
            print(column, ":", count_null, ",", percent_null,"%")
else:
    print("the dataset has not null values")
```

```python
# 10) print distinct languages which there are in the "LanguageWorkedWith" column

print("number of the languages present is:",
      df.LanguageWorkedWith.nunique())
```
